```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, TimeDistributed
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# -------------------------------
# 1. Sample Dataset
# -------------------------------
sentences = [
    ["I", "love", "NLP"],
    ["NLP", "is", "fun"],
    ["I", "enjoy", "learning"]
]

labels = [
    ["PRON", "VERB", "NOUN"],
    ["NOUN", "VERB", "ADJ"],
    ["PRON", "VERB", "VERB"]
]

# -------------------------------
# 2. Build Vocabulary
# -------------------------------
words = list(set(word for sent in sentences for word in sent))
word2idx = {w: i+1 for i, w in enumerate(words)}  # 0 reserved for padding

tags = list(set(tag for lab in labels for tag in lab))
tag2idx = {t: i for i, t in enumerate(tags)}

# -------------------------------
# 3. Convert to Sequences
# -------------------------------
X = [[word2idx[w] for w in sent] for sent in sentences]
y = [[tag2idx[t] for t in lab] for lab in labels]

# Padding
max_len = max(len(s) for s in X)
X = pad_sequences(X, maxlen=max_len, padding='post')
y = pad_sequences(y, maxlen=max_len, padding='post')

# Reshape y for training
y = np.expand_dims(y, -1)

# -------------------------------
# 4. Build LSTM Model
# -------------------------------
model = Sequential([
    Embedding(input_dim=len(word2idx)+1, output_dim=8, input_length=max_len),
    LSTM(32, return_sequences=True),
    TimeDistributed(Dense(len(tag2idx), activation='softmax'))
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# -------------------------------
# 5. Train
# -------------------------------
model.fit(X, y, epochs=50, verbose=1)

# -------------------------------
# 6. Prediction
# -------------------------------
def predict(sentence):
    seq = [word2idx.get(w, 0) for w in sentence]
    seq = pad_sequences([seq], maxlen=max_len, padding='post')

    pred = model.predict(seq)
    pred = np.argmax(pred, axis=-1)[0]

    idx2tag = {i: t for t, i in tag2idx.items()}

    return [(word, idx2tag[p]) for word, p in zip(sentence, pred[:len(sentence)])]

# -------------------------------
# 7. Test
# -------------------------------
print(predict(["I", "love", "learning"]))
```
