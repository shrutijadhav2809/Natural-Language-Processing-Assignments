```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# -------------------------------
# Sample Data
# -------------------------------
resume = "machine learning data analysis python"

jobs = [
    "data scientist python machine learning",
    "web development html css javascript",
    "artificial intelligence deep learning",
    "accounting finance management"
]

# -------------------------------
# Simple Word Embeddings (mock)
# -------------------------------
# In real case → use Word2Vec / GloVe
embedding_dim = 5

def get_random_vector():
    return np.random.rand(embedding_dim)

# Build embedding dictionary
all_words = set(resume.split())
for job in jobs:
    all_words.update(job.split())

embeddings = {word: get_random_vector() for word in all_words}

# -------------------------------
# Sentence Embedding
# -------------------------------
def sentence_vector(sentence):
    words = sentence.split()
    vectors = [embeddings[word] for word in words if word in embeddings]
    return np.mean(vectors, axis=0)

# -------------------------------
# Compute Similarity
# -------------------------------
resume_vec = sentence_vector(resume)

results = []

for job in jobs:
    job_vec = sentence_vector(job)
    sim = cosine_similarity([resume_vec], [job_vec])[0][0]
    results.append((job, sim))

# Sort results
results = sorted(results, key=lambda x: x[1], reverse=True)

# -------------------------------
# Output
# -------------------------------
print("Job Recommendations:\n")
for job, score in results:
    print(f"{job} → Similarity: {score:.2f}")
```
