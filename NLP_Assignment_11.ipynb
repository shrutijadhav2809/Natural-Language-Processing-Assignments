```python
import nltk
from nltk import word_tokenize, pos_tag
from nltk.chunk import RegexpParser

# Download required resources
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng') # Changed to download 'averaged_perceptron_tagger_eng'

# Sample paragraph
text = """
It was the best of times, it was the worst of times, it was the age of wisdom, it was the age of foolishness, it was the epoch of belief, it was the epoch of incredulity, it was the season of Light, it was the season of Darkness, it was the spring of hope, it was the winter of despair.
"""

# Step 1: Tokenization
words = word_tokenize(text)

# Step 2: POS Tagging
pos_tags = pos_tag(words)

print("\nPOS Tags:\n", pos_tags)

# Step 3: Define Grammar for Noun Phrases
grammar = "NP: {<DT>?<JJ>*<NN.*>+}"

# Step 4: Chunking
chunk_parser = RegexpParser(grammar)
tree = chunk_parser.parse(pos_tags)

print("\nNoun Phrases:\n")

# Step 5: Extract Noun Phrases
for subtree in tree.subtrees():
    if subtree.label() == 'NP':
        np = " ".join(word for word, tag in subtree.leaves())
        print(np)
```
